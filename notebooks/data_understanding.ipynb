**Many cols are categorical but already "encoded" and have numerical type**

```python
df_dogs = df[df['Type'] == 1]

top_breeds_x = df_dogs[['BreedName_x', 'BreedName_y']].value_counts()
total_count = top_breeds_x.sum()

top_breeds_percent_x = (top_breeds_x / total_count) * 100
major_breeds_x = top_breeds_percent_x[top_breeds_percent_x >= 1]
other_breeds_percent_x = top_breeds_percent_x[top_breeds_percent_x < 1].sum()
major_breeds_x['Other'] = other_breeds_percent_x

plt.figure(figsize=(12, 12))
plt.pie(major_breeds_x,  
        labels=major_breeds_x.index,
        wedgeprops={'edgecolor': 'black'},
        shadow=True,
        autopct='%1.1f%%')
plt.title('Dogs breeds percentage')
plt.show()
plt.show()
```

**The majority of the dogs are of mixed breeds**

```python
df_cats = df[df['Type'] == 2]

top_cat_breeds_x = df_cats[['BreedName_x', 'BreedName_y']].value_counts()
total_cat_breeds_x = top_cat_breeds_x.sum()

top_cat_breeds_perc = (top_cat_breeds_x / total_cat_breeds_x) * 100
major_cat_breeds_x = top_cat_breeds_perc[top_cat_breeds_perc >= 1]
other_cat_breeds_x = top_cat_breeds_perc[top_cat_breeds_perc < 1]
major_cat_breeds_x['Other'] = other_cat_breeds_x.sum()

plt.figure(figsize=(12, 12))
plt.pie(major_cat_breeds_x,  
        labels=major_cat_breeds_x.index,
        wedgeprops={'edgecolor': 'black'},
        shadow=True,
        autopct='%1.1f%%')
plt.title('Cat breeds persentage')
plt.show()
```

**The majority of the cats are of domestic breeds**

- Drop the not-numerical data to see correlation matrix

```python
ds = df.drop(['Unnamed: 0', 'Name', 'RescuerID', 
              'Description', 'PetID', 'ColorName_x',
              'ColorName_y', 'ColorName', 'BreedName_x',
              'BreedName_y',  'StateName_x'], axis=1)
ds.columns
ds.info()
```

```python
plt.rcParams["figure.figsize"] = (15,12)
correlation_matrix = ds.corr()
sns.heatmap(
    correlation_matrix,
    vmax=1,
    square=True,
    annot=True,
    fmt='.2f',
    cmap='GnBu',
    cbar_kws={"shrink": .5},
    robust=True
    )
plt.title('Correlation Matrix of features', fontsize=8)
plt.show()
```

**We observe a high level of correlation between the fields 'Dewormed' and 'Vaccinated.' This may indicate that these procedures are often performed together. Therefore, during the Feature Engineering stage, one of these columns should be dropped to improve the quality of modeling.**

- Distribution of health features

```python
fig, axs = plt.subplots(2, 2, figsize=(14, 7))

sns.countplot(x='Vaccinated', data=df, palette='Set2', ax=axs[0, 0])
axs[0, 0].set_title('Count of Vaccinated')
axs[0, 0].grid(True)
axs[0, 0].set_xlabel('Vaccination Status')
axs[0, 0].set_ylabel('Number of Animals')
axs[0, 0].set_xticklabels(['Vaccinated', 'Unvaccinated', 'Not sure'])

sns.countplot(x='Dewormed', data=df, palette='Set2', ax=axs[0, 1])
axs[0, 1].set_title('Count of Dewormed')
axs[0, 1].grid(True)
axs[0, 1].set_xlabel('Dewormation Status')
axs[0, 1].set_ylabel('Number of Animals')
axs[0, 1].set_xticklabels(['Dewormed', 'Undewormed', 'Not sure'])

sns.countplot(x='Sterilized', data=df, palette='Set2', ax=axs[1, 0])
axs[1, 0].set_title('Count of Sterilized')
axs[1, 0].grid(True)
axs[1, 0].set_xlabel('Sterilization Status')
axs[1, 0].set_ylabel('Number of Animals')
axs[1, 0].set_xticklabels(['Sterilized', 'Unsterilized', 'Not sure'])

sns.countplot(x='Health', data=df, palette='Set2', ax=axs[1, 1])
axs[1, 1].set_title('Count of health groups')
axs[1, 1].grid(True)
axs[1, 1].set_xlabel('Health Status')
axs[1, 1].set_ylabel('Number of Animals')
axs[1, 1].set_xticklabels(['Healthy', 'Minor Injury', 'Serious Injury'])

plt.tight_layout()
plt.show()
```

- Перевіряю вплив стерилізації на швидкість адопції

```python
# Create the crosstabs for each of the features, ensuring that NaNs are handled properly
health_adoption = pd.crosstab(df['Health'], df['AdoptionSpeed'])
sterilized_adoption = pd.crosstab(df['Sterilized'], df['AdoptionSpeed'])
dewormed_adoption = pd.crosstab(df['Dewormed'], df['AdoptionSpeed'])
vaccinated_adoption = pd.crosstab(df['Vaccinated'], df['AdoptionSpeed'])

# Plot each crosstab on its own subplot
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

health_adoption.plot(kind='bar', stacked=False, ax=axes[0, 0], color=['lightblue', 'lightgreen', 'salmon', 'lightcoral', 'gold'])
axes[0, 0].set_title('AdoptionSpeed vs Health')
axes[0, 0].set_xlabel('Health')
axes[0, 0].set_ylabel('Number of Pets')
axes[0, 0].legend(title='AdoptionSpeed')

sterilized_adoption.plot(kind='bar', stacked=False, ax=axes[0, 1], color=['lightblue', 'lightgreen', 'salmon', 'lightcoral', 'gold'])
axes[0, 1].set_title('AdoptionSpeed vs Sterilized')
axes[0, 1].set_xlabel('Sterilized')
axes[0, 1].set_ylabel('Number of Pets')
axes[0, 1].legend(title='AdoptionSpeed')

dewormed_adoption.plot(kind='bar', stacked=False, ax=axes[1, 0], color=['lightblue', 'lightgreen', 'salmon', 'lightcoral', 'gold'])
axes[1, 0].set_title('AdoptionSpeed vs Dewormed')
axes[1, 0].set_xlabel('Dewormed')
axes[1, 0].set_ylabel('Number of Pets')
axes[1, 0].legend(title='AdoptionSpeed')

vaccinated_adoption.plot(kind='bar', stacked=False, ax=axes[1, 1], color=['lightblue', 'lightgreen', 'salmon', 'lightcoral', 'gold'])
axes[1, 1].set_title('AdoptionSpeed vs Vaccinated')
axes[1, 1].set_xlabel('Vaccinated')
axes[1, 1].set_ylabel('Number of Pets')
axes[1, 1].legend(title='AdoptionSpeed')

# Adjust layout to make sure the subplots don't overlap
plt.tight_layout()
plt.show()
```

- Перевірка розподілів швидкості адопції для різних груп стерилізації

```python
# Define the features to plot
features = ['Sterilized', 'Health', 'Dewormed', 'Vaccinated']

# Create a figure with 2x2 subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Loop through each feature to create a boxplot
for i, feature in enumerate(features):
    ax = axes[i // 2, i % 2]  # Determine position for subplot
    sns.boxplot(x=feature, y='AdoptionSpeed', data=df, ax=ax)
    ax.set_title(f'AdoptionSpeed vs {feature}')
    ax.set_xlabel(f'{feature} (1 - Yes, 2 - No, 3 - Unknown)')
    ax.set_ylabel('AdoptionSpeed')

# Adjust the layout so the plots don't overlap
plt.tight_layout()
plt.show()
```

**It seems like pets that worse from the point of view of prevention have higher chance of rapid adoption :)**
